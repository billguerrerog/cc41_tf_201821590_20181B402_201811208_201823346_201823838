```python
import pandas as pd
import random as r
from google.colab import files

import matplotlib.pyplot as plt
import numpy as np

import math as m
```

```python
data_alm = []

while len(data_alm) < 100:
    x = r.randint(0, 999)
    y = r.randint(0, 999)
    
    new_coord = [x, y]
    if new_coord not in data_alm:
        data_alm.append(new_coord)

df = pd.DataFrame(data_alm, columns= ['Coord X', 'Coord Y'])
df.to_csv(r'data_alm.csv')
#files.download("data_alm.csv")
df
```

```python
data_ent = []

while len(data_ent) < 5000:
    x = r.randint(0, 999)
    y = r.randint(0, 999)
    
    new_coord = [x, y]
    if new_coord not in data_ent and new_coord not in data_alm:
        data_ent.append(new_coord)

df2 = pd.DataFrame(data_ent, columns= ['Coord X', 'Coord Y'])
df2.to_csv(r'data_ent.csv')
#files.download("data_ent.csv")
df2
```

```python
df3 = pd.DataFrame()
data = list()
cont, cont2 = 0, 0

for x1, y1 in data_ent:
    cont2 = 0
    for x2, y2 in data_alm:
      data.append(m.sqrt(pow(x2-x1,2)+pow((y2 - y1), 2)))
      cont2 += 1
    df3[cont] = data
    cont += 1
    data.clear()

df3.to_csv(r'data_cost.csv')
#files.download("data_cost.csv")
df3
```

```python
for x,y in data_ent:
    plt.plot(x, y, '.', color='blue')
for x,y in data_alm:
    plt.plot(x, y, '.', color='red')
cm = 1/2.54
plt.rcParams["figure.figsize"] = [60*cm, 30*cm]
plt.rcParams["figure.autolayout"] = True
plt.show()
```
